```python
meal_cost = 12
tip_percent = 20
tax_percent = 8
def solve(meal_cost, tip_percent, tax_percent):
    tip = meal_cost * tip_percent/100
    tax = meal_cost * tax_percent/100
    total_cost = meal_cost + tip + tax
    print(int(total_cost + 0.5))
solve(meal_cost, tip_percent, tax_percent)
```
